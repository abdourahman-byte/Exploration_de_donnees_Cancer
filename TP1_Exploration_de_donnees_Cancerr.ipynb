# 10. Analyse Univariée

## Variable : diagnosis(1=m, 0=b)

### - vérification du type de distribution

```python
# Renommer la varible dépendante
cancer = cancer.rename(columns={"diagnosis(1=m, 0=b)": "diagnosis"})
```

```python
total = cancer.diagnosis.value_counts()
percent1 = cancer.diagnosis.value_counts()/len(cancer)*100
percent2 = (round(percent1, 2))
diagnosis = pd.concat([total, percent2], axis=1, keys=['Total','%'])
diagnosis
```

```python
# regrouper les données en fonction de la variable d'intérêt

cancer.groupby(['diagnosis']).mean().T
```
